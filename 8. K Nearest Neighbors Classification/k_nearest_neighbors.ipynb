# K-Nearest Neighbors (K-NN)

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
dataset = pd.read_csv(io.BytesIO(uploaded['Social_Network_Ads_Previous.csv']), delimiter=";")
```

```python
dataset.head()
```

```python
#dataset = pd.read_csv('Social_Network_Ads.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

## Splitting the dataset into the Training set and Test set

```python
# Import the train_test_split function from the scikit-learn library
from sklearn.model_selection import train_test_split

# Split the dataset into training and testing sets
# X is the feature matrix, y is the target variable
# test_size = 0.2 indicates that 20% of the data will be used for testing
# random_state = 0 ensures reproducibility by fixing the random seed
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

## Feature Scaling

```python
# Import the StandardScaler class from the scikit-learn library
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler object
sc = StandardScaler()

# Fit the scaler to the training data and transform it
# This scales and centers the features in X_train
X_train = sc.fit_transform(X_train)

# Transform the test data using the same scaler
# This ensures consistency in scaling between training and testing sets
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

## Training the K-NN model on the Training set

```python
# Import the KNeighborsClassifier from scikit-learn
from sklearn.neighbors import KNeighborsClassifier

# Create a KNeighborsClassifier with specified parameters
# n_neighbors is the number of neighbors to consider (in this case, 5)
# metric is the distance metric used for calculating the distance between points
# Different metrics can be chosen, such as 'euclidean', 'manhattan', 'cityblock', 'cosine', 'haversine', etc.

# Euclidean distance
classifier = KNeighborsClassifier(n_neighbors=5, metric='euclidean')

# Manhattan distance
# classifier = KNeighborsClassifier(n_neighbors=5, metric='manhattan')

# Cityblock distance
# classifier = KNeighborsClassifier(n_neighbors=5, metric='cityblock')

# Cosine similarity
# classifier = KNeighborsClassifier(n_neighbors=5, metric='cosine')

# Haversine distance (used for geographical data, typically latitude and longitude)
# classifier = KNeighborsClassifier(n_neighbors=5, metric='haversine')

# Fit the classifier to the training data
classifier.fit(X_train, y_train)
```

## Predicting the Test set results

```python
# Predictions from the KNN classifier
y_pred = classifier.predict(X_test)
```

## Making the Confusion Matrix

```python
from sklearn.metrics import accuracy_score

# Evaluate accuracy
accuracy_knn = accuracy_score(y_test, y_pred)
print("KNN Accuracy:", accuracy_knn)
```

```python
# Import the confusion_matrix function from the scikit-learn library
from sklearn.metrics import confusion_matrix

# Calculate the confusion matrix to evaluate the performance of the classifier
# y_test contains the true labels, and y_pred contains the predicted labels
cm = confusion_matrix(y_test, y_pred)

# Print the confusion matrix
print("Confusion Matrix:")
print(cm)
```

```python
# Import the necessary libraries for visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Use Seaborn to create a heatmap of the confusion matrix
# annot=True adds the numerical values in each cell, fmt="d" formats them as integers
# cmap="Blues" sets the color map to shades of blue
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")

# Set the title, xlabel, and ylabel for the plot
plt.title("Confusion Matrix")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")

# Display the plot
plt.show()
```
